# Building a Logistic Regression model
## Splitting the dataset into training set and test set

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop('Class',axis=1)
y = df['Class']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

## Training and predicting

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(max_iter=1000)
```

```python
lr.fit(X_train,y_train)
prediction = lr.predict(X_test)
```

## Evaluation

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, recall_score_score
```

```python
accuracy_score(y_test,prediction)
```

Looks great right? We created a model with 99% accuracy. But look at this.

```python
print(confusion_matrix(y_test,prediction))
print('\n')
print(classification_report(y_test,prediction))
```

Our model predicted 100% the safe cases but not the fraud one. There are 56 fraud cases that were being classified as safe (false negative). So, we have to develop a model that can detect the fraud cases. Accuracy is not the right metrics but instead we can use recall, precision, roc curve, or precision-recall-curve.

How do we improve our model, given that the recall score of our model is just 62%. There are two ways, either oversampling or undersampling. Oversampling is where we duplicate the minority class until it has the same instances as the majority class. Undersampling is where we only use few data points from the majority class. The problem with oversampling is it can cause overfitting to the minority class. On the other hand, undersampling can cause loss of information.

## Oversampling
I'll be using SMOTE to oversample our data.

```python
from imblearn.over_sampling import SMOTE
```

```python
sm = SMOTE()
```

```python
X_train_new, y_train_new = sm.fit_sample(X_train,y_train.ravel())
```

```python
np.unique(y_train_new, return_counts=True)
```

Now, we have the same amount of positive and negative classes. Let's build our model.

```python
lr = LogisticRegression(max_iter=1000)
```

```python
lr.fit(X_train_new,y_train_new)
```

```python
prediction = lr.predict(X_test)
```

```python
recall_score(y_test,prediction)
```

```python
confusion_matrix(y_test,prediction)
```

By using oversampling technique, we managed to reduce false negatives from 56 to 15.
